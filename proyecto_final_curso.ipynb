```python
# @title **Nombres de variables, su valor mas frecuente y la frecuencia, excepto la variable nombre del estudiante**

def tabla_frecuencias(df):
  """
  Crea una tabla con el valor más frecuente y su frecuencia para cada columna del DataFrame,
  excluyendo el índice, 'NOMBRE DEL ESTUDIANTE' y la numeración.
  """

  tabla = []
  for columna in df.columns:
    if columna not in ['NOMBRE DEL ESTUDIANTE', 'Unnamed: 0']:
      valor_mas_frecuente = df[columna].mode()[0] if not df[columna].mode().empty else None
      frecuencia = df[columna].value_counts().max() if not df[columna].value_counts().empty else None
      tabla.append([columna, valor_mas_frecuente, frecuencia])

  tabla_df = pd.DataFrame(tabla, columns=['Variable', 'Valor más frecuente', 'Frecuencia'])
  return tabla_df

# Aplicar la función al DataFrame df
tabla_frecuencias_df = tabla_frecuencias(df)

# Mostrar la tabla
tabla_frecuencias_df
```

```python
# @title **Rango de los datos**
maximo = max(df['PUNTAJE TOTAL'])
minimo = min(df['PUNTAJE TOTAL'])
Rango = maximo - minimo
print(f"El rango de los datos es: R = {maximo} - {minimo} = {Rango}")
```

```python
# @title **Numero de intervalos con la Regla de Sturges $$k = 1 +3.322log(N)$$**
N = len(df['PUNTAJE TOTAL']) # Total de Datos
k = 1 + 3.322 * np.log10(N) # Número de intervalos
k = int(k)
print(f"El numero de intervalos es: k = 1 + 3.322 * log10({N}) = 1 + 3.322 * log10({N}) = {k}")
```

```python
# @title **Amplitud de cada intervalo $$A = \frac{R}{k}$$**
A = Rango / k # Amplitud de cada intervalo
A = int(A)
print(f"La amplitud de cada intervalo es: A = R / k = {Rango} / {k} = {A}")
```

```python
# @title **Definimos las clases o intervalos a usar**
clases = pd.cut(df['PUNTAJE TOTAL'], k)
clases = clases.value_counts().sort_index()
print(clases)
```

```python
# @title **Crear tabla de frecuencias**
tabla_frecuencia = pd.DataFrame(clases)
tabla_frecuencia
```

```python
# @title **Crear tabla de frecuencias absolutas**
tabla_frecuencia.columns = ['fᵢ']
tabla_frecuencia
```

```python
#@title **Intervalos redondeados**
tabla_frecuencia['Intervalos'] = tabla_frecuencia.index
tabla_frecuencia['Intervalos'] = tabla_frecuencia['Intervalos'].apply(lambda x: f'[{round(x.left,1)}, {round(x.right,1)})')
print(tabla_frecuencia)
```

```python
# @title **Cerrando el último intervalo**
ultimo_intervalo = tabla_frecuencia.index[-1]

# Aplicar el formato adecuado para cada intervalo
tabla_frecuencia['Intervalos'] = tabla_frecuencia.index
tabla_frecuencia['Intervalos'] = tabla_frecuencia['Intervalos'].apply(
    lambda x: f'[{round(x.left,1)}, {round(x.right,1)}]' if x == ultimo_intervalo else f'[{round(x.left,1)}, {round(x.right,1)})'
)

print(tabla_frecuencia)
```

```python
# @title **Calcular la marca de clase (Xᵢ)**
tabla_frecuencia['Xᵢ'] = [round(interval.mid,1) for interval in tabla_frecuencia.index]
print(tabla_frecuencia)
```

```python
# @title **Calcular la frecuencia absoluta acumulada (Fᵢ)**
tabla_frecuencia['F acumulada (Fᵢ)'] = tabla_frecuencia['fᵢ'].cumsum()
print(tabla_frecuencia)
```

```python
# @title **Calcular la frecuencia relativa (fᵢ) $f_r = \frac{f_i}{N}$**
tabla_frecuencia['Fr (fᵢ)'] = tabla_frecuencia['fᵢ'] / len(df['PUNTAJE TOTAL'])
print(tabla_frecuencia)
```

```python
# @title **Calcular la frecuencia relativa acumulada ($F_r=F/N$)**
tabla_frecuencia['Fr acumulada (Fᵢ)'] = tabla_frecuencia['Fr (fᵢ)'].cumsum()
tabla_frecuencia
```

```python
# @title **Multiplicar por 100 para mostrar en porcentaje**
# Multiplicar por 100 para mostrar en porcentaje
tabla_frecuencia['Fr (fᵢ)'] = tabla_frecuencia['Fr (fᵢ)'] * 100
tabla_frecuencia['Fr acumulada (Fᵢ)'] = tabla_frecuencia['Fr acumulada (Fᵢ)'] * 100
print(tabla_frecuencia)
```

```python
# @title **Mostrar la tabla final con separación vertical**
print(tabla_frecuencia[['Intervalos', 'fᵢ', 'Xᵢ', 'F acumulada (Fᵢ)', 'Fr (fᵢ)',
                        'Fr acumulada (Fᵢ)']].to_string(index=False))
```

```python
# @title **Calculando la  Media Agrupada $$\bar{X} = \frac{\sum_{i=1}^{k} f_i \cdot X_i}{\sum_{i=1}^{k} f_i}$$**
ptotal=df['PUNTAJE TOTAL']
media_agrupada = np.average(tabla_frecuencia['Xᵢ'], weights=tabla_frecuencia['fᵢ'])
print(f"Media Agrupada: {media_agrupada:.2f}")
print(f"Media individual: {np.mean(ptotal):.2f}")
```

```python
# @title **Calculando la Mediana agrupada (aproximada) $$Me = L_i + \left( \frac{\frac{N}{2} - F_i}{f_i} \right) \cdot c$$**
n = len(ptotal)
tabla_frecuencia['Intervalos'] = tabla_frecuencia.index
F_ant = tabla_frecuencia['fᵢ'].cumsum().shift(1).fillna(0)
Li = tabla_frecuencia['Intervalos'].apply(lambda x: x.left)
marca_clase = tabla_frecuencia['Xᵢ']
f_i = tabla_frecuencia['fᵢ']

# Fórmula de mediana agrupada
L_mediana = Li[(F_ant + f_i).ge(n / 2).idxmax()]
F_ant_mediana = F_ant[(F_ant + f_i).ge(n / 2).idxmax()]
f_mediana = f_i[(F_ant + f_i).ge(n / 2).idxmax()]


# Calcular amplitud de clase correctamente
amplitud_clase = tabla_frecuencia['Intervalos'].apply(lambda x: x.right - x.left).values[0]  # Amplitud del primer intervalo

mediana_agrupada = L_mediana + (n / 2 - F_ant_mediana) / f_mediana * amplitud_clase


print(f"L_i = {L_mediana} es el límite inferior de la clase mediana.")
print(f"N = {n} es el tamaño total de la muestra.")
print(f"F_i = {F_ant_mediana} es la frecuencia acumulada de la clase anterior a la clase mediana.")
print(f"f_i ={f_mediana} es la frecuencia de la clase mediana.")
print(f"c ={amplitud_clase} es el tamaño del intervalo de clase.")
print(f"Mediana agrupada: M = {L_mediana} + ({n / 2} - {F_ant_mediana}) / {f_mediana} * {amplitud_clase }= {mediana_agrupada}")
print(f"Mediana individual: {np.median(ptotal):.2f}")
```

```python
# @title **29. Calculando la moda agrupada $$Mo = L_i + \frac{ f_m - f_{m-1}}{( f_m - f_{m-1})+( f_m - f_{m+1})}*c$$**
indice_clase_modal = f_i.idxmax()  # Índice de la clase con mayor frecuencia
f_m = f_i[indice_clase_modal]  # Frecuencia de la clase modal

# Get the index position of the modal class
modal_index_position = f_i.index.get_loc(indice_clase_modal)

# Access previous and next frequencies using the index position
f_m_1 = f_i.iloc[modal_index_position - 1] if modal_index_position > 0 else 0  # Frecuencia de la clase anterior
f_m_2 = f_i.iloc[modal_index_position + 1] if modal_index_position < len(f_i) - 1 else 0  # Frecuencia de la clase siguiente

L_modal = Li[indice_clase_modal]  # Límite inferior de la clase modal

# Paso 2: Cálculo de la moda agrupada
moda_agrupada = L_modal + ((f_m - f_m_1) / ((f_m - f_m_1) + (f_m - f_m_2))) * amplitud_clase

# Mostrar el resultado
print("****************************************************************************")
print(tabla_frecuencia[['Intervalos', 'fᵢ', 'Xᵢ', 'F acumulada (Fᵢ)', 'Fr (fᵢ)',
                        'Fr acumulada (Fᵢ)']].to_string(index=False))
print("****************************************************************************")
print(f"L_m = {L_modal} es el límite inferior de la clase modal.")
print(f"f_m = {f_m} es la frecuencia de la clase modal.")
print(f"f_m-1 = {f_m_1} es la frecuencia de la clase anterior a la modal.")
print(f"f_m+1 = {f_m_2} es la frecuencia de la clase siguiente a la modal.")
print(f"c = {amplitud_clase} es el tamaño del intervalo de clase.")
print(f"Moda agrupada: Mo = {L_modal} + ({(f_m - f_m_1)} / ({(f_m - f_m_1)} + {(f_m - f_m_2)})) * {amplitud_clase} = {moda_agrupada}")
print(f"Moda individual: {stats.mode(ptotal, keepdims=True).mode[0]}")
```

```python
# @title **Calculando los cuartiles - $$\text{Cuantil} = L + \left( \frac{n \cdot q - F\_ant\_q}{f\_q} \right) \cdot c$$**
# Para cuartiles y deciles agrupados se usa una aproximación similar a la mediana agrupada
def posicion_agrupada(q, n):
    L = Li[(F_ant + f_i).ge(n * q).idxmax()]
    F_ant_q = F_ant[(F_ant + f_i).ge(n * q).idxmax()]
    f_q = f_i[(F_ant + f_i).ge(n * q).idxmax()]
    return L + (n * q - F_ant_q) / f_q * amplitud_clase
```

```python
# @title **Inferencia - Intervalo de confianza del 95% para la media (usando la distribución normal estándar) $$IC(\mu)=\bar{y}_{Definitiva}+/-z_{\alpha/2}\cdot\frac{s}{\sqrt{n}}$$**


# Suponiendo que la desviación estándar poblacional es conocida o que n es grande (n>=30)
# Si no se conoce la desviación estándar poblacional se usa la distribución t-student.
# Si n>=30, las distribuciones t-student y normal estandar son muy parecidas

# Media y desviación estándar de los datos individuales
media = np.mean(ptotal)
desviacion_estandar = np.std(ptotal, ddof=1)  # Desviación estándar muestral


# Nivel de confianza
nivel_confianza = 0.95

# Calcula el valor z para el nivel de confianza deseado (ej. 95% de confianza)
z = stats.norm.ppf(1 - (1 - nivel_confianza) / 2) # El valor z asociado a la cola superior 0.025 es 1.96 para un intervalo de confianza del 95%

# Error estándar
n = len(ptotal)
error_estandar = desviacion_estandar / np.sqrt(n)

# Calcula el margen de error
margen_error = z * error_estandar

# Calcula el intervalo de confianza
limite_inferior = media - margen_error
limite_superior = media + margen_error

# Resultados
print("****************************************************************************")
print(tabla_frecuencia[['Intervalos', 'fᵢ', 'Xᵢ', 'F acumulada (Fᵢ)', 'Fr (fᵢ)',
                        'Fr acumulada (Fᵢ)']].to_string(index=False))
print("****************************************************************************")
print(f"Intervalo de confianza del 95% para la media (usando la distribución normal estándar):")
print(f"({round(limite_inferior, 2)}, {round(limite_superior, 2)})")
```

# **Prueba de hipótesis**

**Ejemplo:**
Probar la hipótesis de si existe una diferencia significativa en el rendimiento académico entre hombres y mujeres

Para probar la hipótesis utilizaremos una **prueba t de dos muestras independientes** para comparar las medias de las notas entre ambos grupos.

### **Hipótesis:**
- **Hipótesis nula (H₀):** No hay diferencia significativa en el rendimiento académico entre hombres y mujeres (las medias son iguales).
- **Hipótesis alternativa (H₁):** Hay una diferencia significativa en el rendimiento académico entre hombres y mujeres (las medias no son iguales).

### **Pasos para la prueba:**
1. **Visualización inicial** mediante un gráfico de boxplot para comparar las distribuciones de notas entre hombres y mujeres.
2. **Cálculo de la media** de las notas para cada grupo (hombres y mujeres).
3. **Prueba t de Student** para determinar si la diferencia en las medias es estadísticamente significativa.






**1. Visualización inicial**


```python
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Visualización inicial: Boxplot para comparar hombres y mujeres en términos de notas
plt.figure(figsize=(5, 4))
sns.boxplot(x='SEXO', y='PUNTAJE TOTAL', data=df)
plt.title('Comparación del Rendimiento Académico entre Hombres y Mujeres')
plt.xlabel('Sexo')
plt.ylabel('Puntaje Total')
plt.grid(True)
plt.show()


```

El gráfico de boxplot muestra la distribución de los puntajes Ttales para hombres y mujeres. Se observa que hay una ligera diferencia en las distribuciones, pero no es muy marcada.


### **2. Cálculo de la media de notas**


```python
# 2. Cálculo de la media de notas por sexo
media_notas_hombres = df[df['SEXO'] == 'M']['PUNTAJE TOTAL'].mean()
media_notas_mujeres = df[df['SEXO'] == 'F']['PUNTAJE TOTAL'].mean()

Media_Notas_Hombres_y_Mujeres=np.round((media_notas_hombres, media_notas_mujeres),2) # uso de tuplas para conservar los datos
tuple(Media_Notas_Hombres_y_Mujeres)  # imprimir valores con 2 decimales
```

- **Media de notas para hombres:** 237.2
- **Media de notas para mujeres:** 226.26

### **3. Prueba t de Student**

```python
import scipy.stats as stats # Import the scipy.stats module

# 3. Prueba t de Student para comparar medias entre hombres y mujeres
hombres_notas = df[df['SEXO'] == 'M']['PUNTAJE TOTAL']
mujeres_notas = df[df['SEXO'] == 'F']['PUNTAJE TOTAL']

t_stat_sexo, p_value_sexo = stats.ttest_ind(hombres_notas, mujeres_notas)

# Resultados
(t_stat_sexo, p_value_sexo)
```

- **Estadístico t:** 0.8
- **Valor p:** 0.4

**Interpretación:** La prueba t sugiere que no hay una diferencia estadísticamente significativa en las medias de las notas entre hombres y mujeres, ya que el valor p (0.4) es mayor que 0.05.

### **Conclusión**
No hay evidencia suficiente para rechazar la hipótesis nula (H₀), lo que indica que no hay una diferencia significativa en el rendimiento académico entre hombres y mujeres en este conjunto de datos. Aunque la media de los hombres es ligeramente superior a la de las mujeres, esta diferencia no es estadísticamente significativa.
